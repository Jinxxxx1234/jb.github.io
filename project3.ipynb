## 2. Cleaning & Feature Engineering

Typical NYC yellow taxi data includes columns like:
- `tpep_pickup_datetime`, `tpep_dropoff_datetime`
- `passenger_count`
- `trip_distance`
- `fare_amount`, `tip_amount`, `total_amount`
- Pickup / dropoff location IDs

We will:
1. Standardize column names (lower-case, strip spaces)
2. Convert pickup and dropoff times to `datetime`
3. Convert key numeric columns (distance, fares, passengers)
4. Create new features:
   - trip duration (minutes)
   - pickup hour of day
   - pickup weekday
   - pickup date

```python
# Standardize column names
df.columns = [c.strip().lower() for c in df.columns]
print('Standardized columns:')
print(df.columns.tolist())

# Helper to find a column by possible names
def find_col(possible_names):
    for name in df.columns:
        for cand in possible_names:
            if name == cand.lower():
                return name
    return None

pickup_col = find_col(['tpep_pickup_datetime', 'pickup_datetime', 'pickup_time'])
dropoff_col = find_col(['tpep_dropoff_datetime', 'dropoff_datetime', 'dropoff_time'])
passenger_col = find_col(['passenger_count'])
dist_col = find_col(['trip_distance'])
fare_col = find_col(['fare_amount'])
tip_col = find_col(['tip_amount'])
total_col = find_col(['total_amount'])

print('\nDetected key columns:')
for label, col in [
    ('pickup', pickup_col),
    ('dropoff', dropoff_col),
    ('passenger_count', passenger_col),
    ('trip_distance', dist_col),
    ('fare_amount', fare_col),
    ('tip_amount', tip_col),
    ('total_amount', total_col),
]:
    print(f'{label:15s}: {col}')
```

```python
# Convert datetime columns
if pickup_col is not None:
    df[pickup_col] = pd.to_datetime(df[pickup_col], errors='coerce')
if dropoff_col is not None:
    df[dropoff_col] = pd.to_datetime(df[dropoff_col], errors='coerce')

# Convert numeric columns
numeric_candidates = [passenger_col, dist_col, fare_col, tip_col, total_col]
for col in numeric_candidates:
    if col is not None:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Create engineered time features
if pickup_col is not None:
    df['pickup_datetime'] = df[pickup_col]
    df['pickup_date'] = df[pickup_col].dt.date
    df['pickup_hour'] = df[pickup_col].dt.hour
    df['pickup_weekday'] = df[pickup_col].dt.dayofweek  # 0=Mon, 6=Sun

# Trip duration in minutes
if (pickup_col is not None) and (dropoff_col is not None):
    df['trip_duration_min'] = (df[dropoff_col] - df[pickup_col]).dt.total_seconds() / 60

df.head()
```

### 2.1 Basic Sanity Filters

Here we can apply simple filters to remove clearly invalid trips, such as:
- Non-positive or extremely large distances
- Non-positive or extremely long trip durations
- Negative fares

```python
df_clean = df.copy()

# Filter on distance (e.g., 0 < distance < 100 miles)
if dist_col is not None:
    df_clean = df_clean[(df_clean[dist_col] > 0) & (df_clean[dist_col] < 100)]

# Filter on duration (e.g., 0 < duration < 240 minutes)
if 'trip_duration_min' in df_clean.columns:
    df_clean = df_clean[(df_clean['trip_duration_min'] > 0) & (df_clean['trip_duration_min'] < 240)]

# Filter on fare (e.g., fare > 0)
if fare_col is not None:
    df_clean = df_clean[df_clean[fare_col] > 0]

print('Original rows:', len(df))
print('Rows after basic filters:', len(df_clean))
```

## 3. Exploratory Data Analysis (EDA)

### 3.1 Distributions of Key Numeric Variables

We look at basic distributions:
- Trip distance
- Trip duration (if available)
- Fare amount
- Tip and total amounts

```python
def px_hist(col, nbins=50, title=None):
    if col is None or col not in df_clean.columns:
        print(f'Skipping {col}: not found.')
        return
    if title is None:
        title = f'Distribution of {col}'
    fig = px.histogram(df_clean, x=col, nbins=nbins, title=title)
    fig.show()
    HTML(fig.to_html(include_plotlyjs="cdn", full_html=False))

px_hist(dist_col, nbins=50, title=f'Distribution of {dist_col}')
if 'trip_duration_min' in df_clean.columns:
    px_hist('trip_duration_min', nbins=50, title='Distribution of Trip Duration (min)')
px_hist(fare_col, nbins=50, title='Distribution of Fare Amount')
px_hist(tip_col, nbins=50, title='Distribution of Tip Amount')
px_hist(total_col, nbins=50, title='Distribution of Total Amount')
```

All these histograms are positively skewed which totally matches the real-life common sense that short taxi trips are more common (expecially during midnight when trains are out of service).

### 3.2 Temporal Patterns

We examine how demand (number of trips) changes by:
- Hour of day
- Day of week

This is useful for understanding rush hours and weekday vs weekend patterns.

```python
# Trips by pickup hour
if 'pickup_hour' in df_clean.columns:
    trips_by_hour = df_clean.groupby('pickup_hour').size().reset_index(name='count')
    fig = px.bar(trips_by_hour, x='pickup_hour', y='count',
                 title='Trips by Pickup Hour')
    fig.show()
    HTML(fig.to_html(include_plotlyjs="cdn", full_html=False))

# Trips by weekday
if 'pickup_weekday' in df_clean.columns:
    trips_by_wd = df_clean.groupby('pickup_weekday').size().reset_index(name='count')
    fig = px.bar(trips_by_wd, x='pickup_weekday', y='count',
                 title='Trips by Weekday (0=Mon, 6=Sun)')
    fig.show()
    HTML(fig.to_html(include_plotlyjs="cdn", full_html=False))
```

As we can derive from the two histograms that the data only includes a short time window over the New Year's Eve in 2023 (from 2022.12.31 to 2023.1.1), despite it's unable to figure out what time of the day or what day of the week that we have the most passengers, we could still use this data to see:
1. When did the number of order started to surge?
2. Whether the passengers tend to tip more after midnight?

To answer the two questions we could use the following codes:

```python
df_clean['pickup_minute'] = df_clean[pickup_col].dt.floor('T')

minute_counts = df_clean.groupby('pickup_minute').size().reset_index(name='count')

fig = px.line(minute_counts, x='pickup_minute', y='count',
              title='Trips per Minute (High-Resolution Demand)')
fig.show()
HTML(fig.to_html(include_plotlyjs="cdn", full_html=False))
```

```python
df_clean['tip_rate'] = df_clean[tip_col] / df_clean[fare_col]

fig = px.histogram(df_clean, x='tip_rate', nbins=40,
                   title='Distribution of Tip Rate')
fig.show()
HTML(fig.to_html(include_plotlyjs="cdn", full_html=False))
```

```python
df_clean['is_after_midnight'] = df_clean[pickup_col].dt.hour >= 1

fig = px.box(df_clean, x='is_after_midnight', y='tip_rate',
             labels={'is_after_midnight': 'After Midnight'},
             title='Tip Rate: Before vs After Midnight')
fig.show()
HTML(fig.to_html(include_plotlyjs="cdn", full_html=False))
```

## A Small Sum Up
1. The minute-level pickup distribution exhibits a clear event-driven pattern: a pre-midnight buildup, an abrupt dip at exactly midnight, and a strong post-celebration rebound. This behavior is highly characteristic of large-scale public events such as New Years Eve in NYC, where mobility freezes at the celebration peak and resumes sharply as crowds begin dispersing.
2. Tip rates follow a positively skewed distribution with a dominant cluster between 10% and 30%, consistent with normative tipping conventions in NYC. A small number of outliers exceeding 100% likely reflect celebratory generosity or short-fare trips where fixed-amount tips inflate the percentage.
3. A clear behavioral shift emerges across midnight: passengers before midnight tip both more generously and with greater variability, while post-midnight tipping becomes more conservative and tightly distributed. This pattern aligns with consumer psychology research showing elevated prosocial spending during emotionally salient moments (such as the NYE countdown), followed by a reversion to more utilitarian decision-making once the event concludes.

### 3.3 Relationship between Distance and Fare

We expect longer trips to cost more. Here we visualize the relationship between
trip distance and fare amount using an interactive scatter plot.

```python
if (dist_col is not None) and (fare_col is not None):
    fig = px.scatter(df_clean, x=dist_col, y=fare_col, opacity=0.3,
                     title='Fare Amount vs Trip Distance')
    fig.show()
    HTML(fig.to_html(include_plotlyjs="cdn", full_html=False))

    corr = df_clean[[dist_col, fare_col]].corr().iloc[0, 1]
    print(f'Correlation between distance and fare: {corr:.3f}')
```

The scatter plot shows a very strong positive correlation between distance and fare (r  0.955):
		Base fare + distance-based + time-based components produce a near-linear relationship.
		Deviations above the trend reflect congestion or prolonged idling, which is expected on NYE.
		Very long-distance trips, though fewer, remain consistent with the meter structure.
